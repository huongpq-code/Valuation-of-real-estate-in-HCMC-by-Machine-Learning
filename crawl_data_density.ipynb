```python
import pandas as pd # library for data analysis
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML documents
```

```python
# get the response in the form of html
wikiurl="https://vi.wikipedia.org/wiki/Danh_s%C3%A1ch_%C4%91%C6%A1n_v%E1%BB%8B_h%C3%A0nh_ch%C3%ADnh_c%E1%BA%A5p_huy%E1%BB%87n_c%E1%BB%A7a_Vi%E1%BB%87t_Nam"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
indiatable=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(indiatable))
# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head())
```

```python
density = df.drop(["Số thứ tự", "Chú thích", "Số ĐVHC cấp xã"], axis=1)
density = density.rename(columns={"Tên đơn vị hành chính": "district_name","Tỉnh/Thành phố": "city", "Dân số (người)": "population", "Diện tích (km²)": "lot_area", "Mật độ dân số (người/km²)": "density"})
```

```python
density = density[density['city'] == 'Thành phố Hồ Chí Minh']
print(density.to_string())
```
